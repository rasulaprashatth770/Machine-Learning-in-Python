```python
book_data = book.pivot_table(index='userid',
                   columns='booktitle',
                   values='bookrating').reset_index(drop=True)
book_data
```

```python
book_data.fillna(0,inplace=True)
book_data
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
user_sim=1-pairwise_distances(book_data.values,metric='cosine')
user_sim
```

```python
user_sim_data=pd.DataFrame(user_sim)
user_sim_data
```
