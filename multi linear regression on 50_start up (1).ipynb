```python
import numpy as np
import pandas as pd
df = pd.read_csv("C:/Users/rasul/Downloads/50_Startups (2).csv")
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["R&D Spend"],df["Profit"])
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.show()

import matplotlib.pyplot as plt
plt.scatter(df["Administration"],df["Profit"])
plt.xlabel("Administration")
plt.ylabel("Profit")
plt.show()

import matplotlib.pyplot as plt
plt.scatter(df["Marketing Spend"],df["Profit"])
plt.xlabel("Marketing Spend")
plt.ylabel("Profit")
plt.show()

import matplotlib.pyplot as plt
plt.scatter(df["State"],df["Profit"])
plt.xlabel("State")
plt.ylabel("Profit")
plt.show()
```

```python
df.corr()
```

```python
X = df[["R&D Spend"]] # M1
Y = df["Profit"]
```

```python
import statsmodels.formula.api as smf
model = smf.ols('Profit~X',data=df).fit()
model.summary()
```

```python
model.rsquared.round(4)
model.rsquared_adj.round(4)
Y_pred = model.predict(X)

# step5: calculating the error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y,Y_pred)
print("Mean squarred error: ",mse.round(2))
print("Root Mean squarred error: ",np.sqrt(mse).round(2))
```
