```python
#data transformation
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x_scale=ss.fit_transform(x)
x_scale
```

```python
#data partition
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scale,y,stratify=y)
```

```python
#model fitting
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5,p=2) #k=5 p=2 ecledian distance
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred_train=knn.predict(x_train)
y_pred_test=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Training accuracy",accuracy_score(y_train,y_pred_train).round(4))
print("Test accuracy",accuracy_score(y_test,y_pred_test).round(4))
```
